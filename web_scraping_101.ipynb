##Web Scraping

```python
# Texto de https://www.google.com/recaptcha/about/
import requests, bs4
```

```python
res = requests.get('https://www.google.com/recaptcha/about/')
soup = bs4.BeautifulSoup(res.text, 'html.parser')
x = soup.find_all('div')
text = ''
for i in x:
    text = text + i.text.strip() + '\n'

print(text)
```

```python
#Mostrar texto de etiquetas a
res = requests.get('https://www.google.com/recaptcha/about/')
soup = bs4.BeautifulSoup(res.text, 'html.parser')
z = soup.find_all('a')
text = ''
for i in z:
    text = text + i.text.strip() + '\n'

print(text)
```

```python
#Encontrar todos los links
a = soup.find_all('a')
text = ''

#Hacemos un bucle for para buscar el texto (links en este caso)
for i in a:
    text = text + i.get('href') + '\n'

print(text)
```
