###Análisis de los resultados de un formulario importado en .xlsx

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
data = "/content/drive/MyDrive/Analisis_Bootcamp/Formulario sin título (Respuestas).xlsx"
df = pd.read_excel(data)
df.head()
```

```python
df.columns
```

```python
df.columns = ['Time', 'gender',
       'age', 'nivel_academico',
       'lugar_residencia',
       'Stem_antes',
       'Interes_carreras',
       'informado_intereses',
       'molestia_estudio_sexo_opuesto',
       'ansiedad_al_estudiarSTEM',
       'soledad_academico',
       '[Mapas con universidades en tu ubicación]',
       '[Ranking de universidades para cada carrera de STEM]',
       '[Información de carreras STEM existentes]',
       '[Testimonios de mujeres en STEM]',
       '[Simulador de postulación a carreras STEM y becas]',
       '[Comunidades entre mujeres]',
       '[Tener un usuario propio para ingresar]']
```

```python
plt.figure(figsize=(15, 8))
sns.countplot(data=df, x='nivel_academico')
plt.title('Nivel Académico')
plt.xticks(rotation=45)
plt.show()
```

#Uso de colores

```python
plt.figure(figsize=(15, 8))
colores_personalizados = ["#C49DE6", "#FF9770", "green", "purple"]
sns.countplot(data=df, x='Stem_antes', palette=colores_personalizados)
plt.title('Había escuchado sobre STEM antes')

plt.show()
```

```python
plt.figure(figsize=(15, 8))
colores_personalizados = ["#C49DE6", "#FF9770", "green", "purple"]
sns.countplot(data=df, x='molestia_estudio_sexo_opuesto', palette=colores_personalizados)
plt.title('Había escuchado sobre STEM antes')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
colores_personalizados = ["#C49DE6", "#FF9770", "#FBFBF2", "#A9EFFF"]
sns.countplot(data=df, x='ansiedad_al_estudiarSTEM', palette=colores_personalizados)
plt.title('Le genera ansiedad estudiar carreras STEM')
plt.show()
```

```python

plt.figure(figsize=(15, 8))
colores_personalizados = ["#C49DE6", "#FF9770","#A9EFFF", "#FBFBF2"]
sns.countplot(data=df, x='soledad_academico', palette=colores_personalizados)
plt.title('Le genera ansiedad estudiar carreras STEM')
plt.show()
```
