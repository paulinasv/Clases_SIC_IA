##Ejemplo RNN según CHAT GPT

```python
#Importamos librerías
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
```

```python
# Datos: secuencia de números del 9 al 0
sequence = np.array([9, 8, 7, 6, 5, 4, 3, 2, 1, 0])
```

```python
# Creamos los datos de entrenamiento
def create_sequences(data, seq_length):
    X = []
    y = []
    for i in range(len(data) - seq_length):
        X.append(data[i:i + seq_length])
        y.append(data[i + seq_length])
    return np.array(X), np.array(y)
```

```python
# Definir longitud de la secuencia
sequence_length = 3  # Por ejemplo, usamos 3 números para predecir el siguiente
X, y = create_sequences(sequence, sequence_length)
```

```python
# Reshape para que sea compatible con RNN (muestras, pasos de tiempo, características)
X = X.reshape((X.shape[0], X.shape[1], 1))
```

```python
# Crear el modelo RNN
model = Sequential()
model.add(SimpleRNN(50, activation='relu', input_shape=(sequence_length, 1)))
model.add(Dense(1))  # Capa de salida con 1 neurona para predecir el siguiente número
```

```python
# Compilar el modelo
model.compile(optimizer='adam', loss='mse')
```

```python
# Entrenar el modelo
model.fit(X, y, epochs=300, verbose=1)
```

```python
# Predecir el siguiente número en la secuencia
test_input = np.array([9, 8, 7])  # Queremos predecir lo que viene después del 7, 8, 9
test_input = test_input.reshape((1, sequence_length, 1))  # Reshape para que el modelo lo entienda
```

```python
prediction = model.predict(test_input, verbose=0)
print(f"Predicción del siguiente número en la secuencia: {prediction}")
```
